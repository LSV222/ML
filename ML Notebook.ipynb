### Measuring our model

#### **Accuracy** - % de casos correctamente predictos del total de predicciones 

Warning! : Con 'class imbalance', es decir, si tenemos clases muy disparejas en distribucion (por ejemplo en clasificacion binaria) , puede ser poco util como métrica.
Ej : 1% de fraude, si mi modelo dice que nada es fraude tiene precision general (accuracy) de 99%
   

  
*MATRIZ DE CONFUSION* 
                  Prediccion =Fraude    Prediccion= No-Fraude
Real= Fraude      Correcto positivo     Falso negativo 
Real= No-Fraude   Falso Positivo        Correcto negativo


Accuracy con matriz de confusion es = Correcto pos + correcto neg / Total              

#### Precision= Correcto positivos / Todas las predicciones de fraude  
_Habla de cuanto de lo que marcamos como fraude, realmente lo es_  
_Una alta precision significa que tenemos bajo falso positivo; es decir marcamos pocas cosas como fraude sin serlo_  
__Va a ser la medida mas importante en modelos que intenten generar marcas para hacer forenses mas precisos__  

#### Recall o Sensibilidad= Correcto positivos / Todo lo realmente fraude  
_Habla de cuanto de lo que es fraude es marcado como tal_  
_Una alta recall habla de alta sensibilidad de deteccion de casos positivos. Es decir que tenemos bajo falso negativo; se nos "escapan" pocas transacciones fraudulentas _  
__Va a ser la medida mas importante en modelos que intenten generar marcas para atrapar a todos los de una clase incluso a un costo alto de confirmación __  

#### F1 score= 2*precison*recall / precision + recall  
_Es la media armonica de precision y recall. Siempre está entre 0 y 1 , y "favorece modelos con precision y recall similares"._  
_Es por tanto mas util si buscamos un modelo que funcione razonablemente bien en ambas metricas_  

__Es una funcion estrictamente creciente con ambas variables. Y con concavidad negativa. La "concavidad parcial" disminuye levemente a valores mayores de la otra varible.__ --Ver Codigo--


```python
# Grafico de F1
import matplotlib.pyplot as plt;from mpl_toolkits.mplot3d import Axes3D;import numpy as np
x = np.arange(0.1, 1.1, 0.1)
y = np.arange(0.1, 1.1, 0.1)

X, Y = np.meshgrid(x, y)
f = 2*X*Y/(X+Y)

fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(X, Y, f)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
