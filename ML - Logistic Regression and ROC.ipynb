### Logistic Regression
La usamos para clasificar. Generalmente haciendo la predicion de categoria en base a si la probabilidad de ser una clase es mayor a 0.5 o no.

Regresion logistica genera un borde lineal de decision (es decir a traves de una combinacion lineal de variables como borde)

El numero que establece cual es el nivel de probablidad se llama "treshold". Con distintos niveles de treshold , obtenemos distintas combinaciones de las metricas de sensibilidad y tasa de falsos positivos (cuantos no son 1 de todos los que predijimos como tales).   
OBS: tambien podemos cambiar algunas de estas metricas por , por ejemplo, precision ( cantidad de casos correctamente predichos dentro de las predicciones de 1 )

En resumen, tendremos una ROC curve por modelo , que nos habla de todos los posibles escenarios de FP , FN y otras metricas paara un modelo; para los distintos treshold. Sirve para elegir escenarios dentro de un mismo modelo y para comparar modelos.
Para comparar modelos, entre otras cosas podemos comparar su area bajo la curva "AUC". Siendo lo mas cercano a 1 los mejores modelos.

p=0 (no exijo que sea probable) es que predigo a todas las obs como 1.  Entonces True positives es 1 y false positives es 1
p=1 (exijo un monton) es que predigo todas las obs como 0. Entonces True positives es 0 y false positives es 0 .
